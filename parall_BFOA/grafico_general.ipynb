```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV
df = pd.read_csv("resultados_30_corridas.csv")  # Aseg√∫rate de usar el nombre correcto del archivo
```

```python
# Graficar Fitness vs Iteraci√≥n
plt.figure(figsize=(10, 6))
for corrida, group in df.groupby(df.index // 3):  # Agrupar por corrida (3 iteraciones por corrida)
    plt.plot(group["Iteraci√≥n"], group["Fitness"], marker="o", label=f"Corrida {corrida + 1}")

plt.xlabel("Iteraci√≥n")
plt.ylabel("Fitness")
plt.title("Fitness vs Iteraci√≥n")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar Tiempo vs Iteraci√≥n
plt.figure(figsize=(10, 6))
for corrida, group in df.groupby(df.index // 3):
    plt.plot(group["Iteraci√≥n"], group["Tiempo"], marker="o", label=f"Corrida {corrida + 1}")

plt.xlabel("Iteraci√≥n")
plt.ylabel("Tiempo (segundos)")
plt.title("Tiempo vs Iteraci√≥n")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar Interacci√≥n vs Iteraci√≥n
plt.figure(figsize=(10, 6))
for corrida, group in df.groupby(df.index // 3):
    plt.plot(group["Iteraci√≥n"], group["Interacci√≥n"], marker="o", label=f"Corrida {corrida + 1}")

plt.xlabel("Iteraci√≥n")
plt.ylabel("Interacci√≥n")
plt.title("Interacci√≥n vs Iteraci√≥n")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Graficar BLOSUM vs Iteraci√≥n
plt.figure(figsize=(10, 6))
for corrida, group in df.groupby(df.index // 3):
    plt.plot(group["Iteraci√≥n"], group["BLOSUM"], marker="o", label=f"Corrida {corrida + 1}")

plt.xlabel("Iteraci√≥n")
plt.ylabel("BLOSUM")
plt.title("BLOSUM vs Iteraci√≥n")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Crear una figura con subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("M√©tricas por Iteraci√≥n y Corrida")

# Fitness vs Iteraci√≥n
for corrida, group in df.groupby(df.index // 3):
    axes[0, 0].plot(group["Iteraci√≥n"], group["Fitness"], marker="o", label=f"Corrida {corrida + 1}")
axes[0, 0].set_title("Fitness vs Iteraci√≥n")
axes[0, 0].set_xlabel("Iteraci√≥n")
axes[0, 0].set_ylabel("Fitness")
axes[0, 0].legend()
axes[0, 0].grid(True)

# Tiempo vs Iteraci√≥n
for corrida, group in df.groupby(df.index // 3):
    axes[0, 1].plot(group["Iteraci√≥n"], group["Tiempo"], marker="o", label=f"Corrida {corrida + 1}")
axes[0, 1].set_title("Tiempo vs Iteraci√≥n")
axes[0, 1].set_xlabel("Iteraci√≥n")
axes[0, 1].set_ylabel("Tiempo (segundos)")
axes[0, 1].legend()
axes[0, 1].grid(True)

# Interacci√≥n vs Iteraci√≥n
for corrida, group in df.groupby(df.index // 3):
    axes[1, 0].plot(group["Iteraci√≥n"], group["Interacci√≥n"], marker="o", label=f"Corrida {corrida + 1}")
axes[1, 0].set_title("Interacci√≥n vs Iteraci√≥n")
axes[1, 0].set_xlabel("Iteraci√≥n")
axes[1, 0].set_ylabel("Interacci√≥n")
axes[1, 0].legend()
axes[1, 0].grid(True)

# BLOSUM vs Iteraci√≥n
for corrida, group in df.groupby(df.index // 3):
    axes[1, 1].plot(group["Iteraci√≥n"], group["BLOSUM"], marker="o", label=f"Corrida {corrida + 1}")
axes[1, 1].set_title("BLOSUM vs Iteraci√≥n")
axes[1, 1].set_xlabel("Iteraci√≥n")
axes[1, 1].set_ylabel("BLOSUM")
axes[1, 1].legend()
axes[1, 1].grid(True)

# Ajustar el espacio entre subplots
plt.tight_layout()
plt.show()
```
